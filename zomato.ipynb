```python
import re
# Python's re module helps find and match patterns in text.
```

```python
def name_clean(text):
    return re.sub(r"[^a-zA-Z0-9 ]", "", text)
data['Name'] = data['Name'].apply(lambda x: name_clean(x))

# re.sub it is use for the finding the given text and fill the altrnative text
```

```python
data.head(10)
```

```python
## removing 'nan', 'NEW', '-' values

# Missing Data Representation: np.nan is used to denote missing or null values in datasets. 
#This is particularly useful in data cleaning and preprocessing.


data["Ratings"]=data["Ratings"].replace("NEW", np.nan)
data['Ratings']=data['Ratings'].replace('NaN',np.nan)
data['Ratings']=data['Ratings'].replace('-',np.nan)
data['Ratings']=data['Ratings'].replace('nan',np.nan)

## function to remove '/5'
def remove_5(value: str):
    if type(value)==str:
        value_new=value.split('/')[0]
        return value_new
    return value
data['Ratings']=data['Ratings'].apply(remove_5)

## converting to float type data

#Data Type Conversion
data['Ratings']=data['Ratings'].astype(float)


print(data['Ratings'].dtypes)
```

```python
# In cost colum contain the interger value with comma like(1,000) this.
```

```python
def cost(value):
    value = str(value)
    if "," in value:
        value = float(value.replace(",",""))
        return value
    else:
        return float(value)
data['Cost_of_2_people'] = data['Cost_of_2_people'].apply(cost)
#print(df['Cost'].head())
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
score=r2_score(Y_test,y_predict)
mae=mean_absolute_error(Y_test,y_predict)
mse=mean_squared_error(Y_test,y_predict)
print("R2 score is ",score)
print("Mean absolute error is ",mae)
print("Mean squared error ",mse)
```

```python
from flask import Flask
from flask import render_template
import joblib
from flask import request
```

```python


app = Flask(__name__)
model = joblib.load('zomato.pkl')

@app.route('/')
def hello(name=None):
    return render_template('index.html', name=name)

@app.route('/result', methods=['GET', 'POST'])
def get_value():
    prediction = None
    if request.method == "POST":
        
        Online_Order = int(request.form.get('Online_Order'))
        Book_Table = int(request.form.get('Book_Table'))
        Votes = int(request.form.get('Votes'))
        Restaurant_Type = int(request.form.get('Restaurant_Type'))
        Dishes_Liked = int(request.form.get('Dishes_Liked'))
        Cuisines = float(request.form.get('Cuisines'))
        Cost_For_2_Person = int(request.form.get('Cost_For_2_Person'))
        Type = int(request.form.get('Type'))
        City = int(request.form.get('City'))

        predict = model.predict([[ Online_Order, Book_Table, Votes, Restaurant_Type, Dishes_Liked, Cuisines, Cost_For_2_Person, Type, City]])
        prediction = predict[0]
    
    return render_template('index.html', prediction=prediction)

if __name__ == '__main__':
    app.run()
```
